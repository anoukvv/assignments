# Assignment 1


## Data

The data are given:


```r
set.seed(123)
df1 <- as.data.frame(matrix(runif(1*50, min = 1, max = 10), ncol = 1)) %>%
    mutate(epsilon = rnorm(50, mean = 0, sd = 2)) %>%
    mutate(response = 3 - 2* V1  + epsilon) %>%
    mutate(group = ifelse(V1 <= 5, 1,2))
```

```r
head(df1)
```

## Asignment 1a

The first assigment is to make boxplot using ggplot with group on the x-axis and V1 on the y-axis.

```r
ggplot(df1,aes(x=group, y=V1, group = group))+
    geom_boxplot()+
    xlab("Group")+
    ylab("V1")
```

I have created a boxplot using ggplot. The data that is used is the dataframe df1. I put the variable 'group' on the x-axis, and the variable 'V1' on the y-axis in the aesthetics layer and later labeled these using xlab and ylab. I indicated that the boxplot should be grouped based on the variable 'group' in the aesthetics layer. 

## Assignment 1b

Run a regression with response variable as a function of V1.
Show the summary statistics of the regression.


```r
reg1 <- lm(data = df1, response ~ V1)
summary(reg1)
```

I have created a linear regression with the lm model. The data used is the dataframe df1. I put 'response' on the y-axis and 'V1'on the x-axis. 

check if the assumptions of linear regression are met with the `gvlma()` function.

```r
gvlma(reg1)
```

Because the decision is 'Assumptions acceptable' in al 5 rows, we may conclude that the assumptions are met. 

## Assignment 1c

Make a scatterplot with: 
* V1 on the x-axis and the response on the y-axis
* Include the regression line in red with confidence interval 
* In a classic theme
* The x-axis should be labeled "Predictor", the y-axis should be labeled ("Response")



```r
ggplot(df1, aes(x=V1, y=response))+
    geom_point()+
    theme_classic()+
    xlab("Predictor")+
    ylab("Response")+
    geom_smooth(method="lm", colour="red")
```

I made a scatterplot with V1 on the x-axis and Response on the y-axis in the aesthetics layer. I then labeled the axes using xlab and ylab. I indicated that the theme shoul be classic using theme_classic(). I added the regression line with confidence interval using geom_smooth and indicated that the colour should be red.

### 2dii

First, create a new dataframe schools1, which is equal to schools.

As you see POSTCODE has a structure of (1234 AB). You should create a new variable PC4 that is equal to the first 4 numbers in POSTCODE

Hint: Use the function `substr()`

Then select the variables PC4 and DENOMINATIE

```r
schools1 <- schools
schools1$PC4 <- substr(schools1$POSTCODE,1,4) 
schools2 <- subset(schools1, select= c("PC4","DENOMINATIE"))
head(schools2)
```

I have added a variable to the schools dataframe, called PC4. ### 2diii

Create the dataframe school_loc as a join from pc4_locations and school1 that combines the columns from both data frames, but only keeps rows where the value in the pc4_locations column matches in both data frames.
And then select the observations with DENOMINATIE is equal to "Rooms-Katholiek" or "Protestants-Christelijk"

```r
school_loc <- merge(x = pc4_locations, y = schools2)
school_loc1 <- subset(school_loc, subset = DENOMINATIE == c("Rooms-Katholiek", "Protestants-Christelijk"))
head(school_loc1)
```

I first merged the dataframes pc4_locations and schools2 into school_loc. Then I subsetted the rows for which the DENOMINATIE variable is equal to 'Rooms-Katholiek' or 'Protestants-Christelijk'.
